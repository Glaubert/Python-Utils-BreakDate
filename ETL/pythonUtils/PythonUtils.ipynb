```python
import pandas as pd
import pandera as pa
```

```python
df = pd.read_csv("VendaServicoGlaubert.csv", sep=";",parse_dates=['DATA_PEDIDO','DATA_SERVICO','DATA_VENCIMENTO','DATA_PAGAMENTO'], dayfirst=True)
df.head(10) 
```

```python
schema = pa.DataFrameSchema(
    columns = {
      "ID_VENDA" : pa.Column(pa.Int),
      "CLIENTE" : pa.Column(pa.String),
      "SERVIÇO" : pa.Column(pa.String),
      "VALOR" : pa.Column(pa.Int),
      "DATA_PEDIDO" : pa.Column(pa.DateTime),
      "DATA_SERVICO" : pa.Column(pa.DateTime),
      "DATA_VENCIMENTO" : pa.Column(pa.DateTime),
      "DATA_PAGAMENTO" : pa.Column(pa.DateTime, nullable=True)
    }
)
```

```python
schema.validate(df)
```

```python
def breakdate(df):
    for column in df:   
        if df[column].dtypes == 'datetime64[ns]':
            df[df[column].name+"_ANO"] = df[df[column].name].dt.year
            df[df[column].name+"_MES"] = df[df[column].name].dt.month
            df[df[column].name+"_DIA"] = df[df[column].name].dt.day
    return df
breakdate(df).head()
```
